```python
data_files = glob.glob('inflammation*.csv')
print(data_files)

def run_text():

    for i in np.arange(0,len(data_files),1):
        if i < len(data_files) -1:
            num = np.loadtxt(data_files[i],delimiter=',')
            num1 = np.loadtxt(data_files[i+1],delimiter=',')
            fig = plt.figure(figsize=(10.0,3.0))
            axis1 = fig.add_subplot(1,3,1)
            axis2 = fig.add_subplot(1,3,2)
            axis3 = fig.add_subplot(1,3,3)
            axis1.plot(np.mean(num,axis=0))
            axis2.plot(np.max(num,axis=0))
            axis3.plot(np.min(num,axis=0))
            axis1.set_ylabel('Mean')
            axis2.set_ylabel('Maximum')
            axis3.set_ylabel('Minimum')
            axis1.set_xlabel('Numbers')
            axis2.set_xlabel('Numbers')
            axis3.set_xlabel('Numbers')
            axis1.set_title(data_files[i])
            axis2.set_title(data_files[i])
            axis3.set_title(data_files[i])
            axis4 = fig.add_subplot(1,3,1)
            axis5 = fig.add_subplot(1,3,2)
            axis6 = fig.add_subplot(1,3,3)
            axis4.plot(np.mean(num,axis=0)-np.mean(num1,axis=0))
            axis5.plot(np.max(num,axis=0)-np.max(num1,axis=0))
            axis6.plot(np.min(num,axis=0)-np.min(num1,axis=0))
            axis4.set_ylabel('Mean')
            axis5.set_ylabel('Maximum')
            axis6.set_ylabel('Minimum')
            axis4.set_xlabel('Numbers')
            axis5.set_xlabel('Numbers')
            axis6.set_xlabel('Numbers')
            axis4.set_title(data_files[i])
            axis5.set_title(data_files[i])
            axis6.set_title(data_files[i])
            plt.tight_layout()
            mn = np.mean(num,axis=0)
            mx = np.max(num,axis=0)
            mini = np.min(num,axis=0)
            mnd = np.mean(num,axis=0)-np.mean(num1,axis=0)
            mxd = np.max(num,axis=0)-np.max(num1,axis=0)
            minid = np.min(num,axis=0)-np.min(num1,axis=0)
            print(data_files[i], mn[35], mx[35], mini[35])
            print(data_files[i]+ '-' +data_files[i+1],mnd[35],mxd[35],minid[35])
```

```python
run_text()
```

```python
def play(value, limit):
    num = np.zeros((1,)) + value
    print(num)
    if num > limit:
        print('Oh, my word!!!!')
    elif num == limit: 
        print('Oh, my universe!!!!')
    else:
        print('Failures are everywhere.')
```

```python
def temp_convert(temp):
    """The input temperature is in units of farenheit."""
    print('Assuming input temperature in is units of Farenheit')
    new_temp_kelvin = ((temp-32)*(5.0/9.0))-273.15
    new_temp_celsius = ((temp-32)*(5.0/9.0))
    print('Kelvin Temperature:',new_temp_kelvin), print('Celsius Temperature',new_temp_celsius)
```

```python
temp_convert(212)
```
